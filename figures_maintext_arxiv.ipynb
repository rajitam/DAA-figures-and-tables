# Loading the data, filtering out poor statistical data

```python
import numpy as np
import scipy 
############## function to filter out poor statistical data ################
def filter(cd_data , control_data, cutoff):
    cd_data = cd_data.T
    control_data = control_data.T
    nbact = cd_data.shape[1]
    nsamp_CD= cd_data.shape[0]
    nsamp_control= control_data.shape[0]
    
    pred =  np.array([[x/x if x > 0 else x for x in cd_data[:,i]] for i in xrange(nbact)])
    pred1 = np.array([np.sum(0.0 + pred[i,:]) for i in xrange(nbact)])/nsamp_CD

    pred =  np.array([[x/x if x > 0 else x for x in control_data[:,i]] for i in xrange(nbact)])
    pred2 = np.array([np.sum(0.0+ pred[i,:]) for i in xrange(nbact)])/nsamp_control

    bad1=[]
    cutoff = 0.38
    for i in xrange(nbact):
        if pred1[i]<cutoff:
            bad1.append(i)
        
    bad2 = []
    for i in xrange(nbact):
        if pred2[i]<cutoff:
            bad2.append(i)

    bad = np.intersect1d(bad1,bad2)
    allbact = np.arange(nbact)

    good = (list(set(allbact) - set(bad)))
    return good
## input to function in the form of data array of shape [ntaxa, nsamples]
def corrected_data(control_data, cd_data, indices):
    nbact = cd_data.shape[0]
    nsamp_CD= cd_data.shape[1]
    nsamp_control= control_data.shape[1]
    CD_data= cd_data.T
    CD_data = 1.0+CD_data ### PSEUDO COUNT #######
    CD_data = np.array([(CD_data[i,:])/sum(CD_data[i,:]) for i in range(nsamp_CD)]) ########## normalization - TSS ##########
    cd_data = np.log(CD_data).T
    
    Control_data = control_data.T
    Control_data = 1.0+Control_data
    Control_data = np.array([(Control_data[i,:])/sum(Control_data[i,:]) for i in range(nsamp_control)]) ########## normalization - TSS ##########
    control_data = np.log(Control_data).T
  
    control= []
    cd = []
    ind1 = np.array(indices).shape[0]
    for i in range(ind1):
        control.append(control_data[indices[i],:])
        cd.append(cd_data[indices[i],:])
 
    return np.array(control), np.array(cd)


```

# Function 1: Host-field transform on the data (Inverts covariance matrix, returns host-field arrays)

```python
###### takes arguments  1,2 --- data array control,cd = (nsamples, ntaxa), 
##### argument 3 --  real number eigenvalue threshold for inversion of covariance matrix
### we use thresh = 0.01
### function returns sorted eigenvalues of covariance matrix, inverted covariance matrix, 
##### transformed host-field arrays for control and cd 

def hfield_transform(control , cd, thresh):
    res = np.concatenate((control,cd), axis = 0)
    cov = np.cov(res.T)

    U,lambd,Vt = scipy.linalg.svd(cov, full_matrices=True, compute_uv=True, overwrite_a=False, check_finite=True)
    x = np.arange(len(lambd))
    for i in range(len(cov)):
        cutoff = len(lambd)
        if lambd[i] < thresh:
            cutoff = i
            break

    trunc = len(lambd)
    lambd_inv = np.zeros(trunc)
    ######### computing inverse singular values #############
    for i in range(cutoff):
        lambd_inv[i] = 1.0/lambd[i] 
    V = Vt.T
    Ut = U.T

    S_inv = scipy.linalg.diagsvd(lambd_inv, trunc, trunc)
    ######## inverse covariance matrix #########
    cov_inv = np.dot(V, np.dot(S_inv, Ut))
    h_con1 = np.dot(cov_inv,control.T).T
    h_cd1 = np.dot(cov_inv,cd.T).T
    return lambd, cov_inv, h_con1, h_cd1

lambd, cov_inv, h_con1, h_cd1 = hfield_transform(control_clus,cd_clus,0.01)
```

# Function 2: performs permutation test on cd and control data (raw or host-field). Returns p-values

```python
#########################################################################################
############################ permutation test ##################

###### takes argument  1, control data array = (nsamples, ntaxa), 
##### argument 2 -- integer number of control samples 
### argument  3, cd data array = (nsamples, ntaxa)
### argument  4, integer number of permutations 
##### function returns array of p-values for ntaxa

def permutation_test(control,nsamp_control,CD,nperm):  
    Control_av = np.mean(control,axis = 0)
    CD_av = np.mean(cd, axis = 0)
    test_stat = np.abs(CD_av - Control_av)  
    resamp = np.concatenate((CD, control), axis = 0)
    ### number of tests ##########
    trunc = len(test_stat)
    k = 0.0+ np.zeros(trunc)
    for j in range(nperm):
        resamp = np.random.permutation(resamp)
        control = resamp[0:nsamp_control,:]
        CD = resamp[nsamp_control::,:]
    
        Control_av = np.mean(control, axis = 0) 
        CD_av = np.mean(CD, axis = 0)
        diff = np.abs(CD_av - Control_av)
    
        for i in range(trunc):
            if diff[i] > test_stat[i]:
                k[i] = k[i]+1.0

    pvals = k/nperm
    return pvals
```

# Function 6: Analyzes synthetic data, accesses multiple copies of control and CD data to compute number of detections as a function of sample sizes (Figure 2A).

```python
#########################################################################################
############################ Analysis of synthetic data for Number of significant detections versus sample size ##################
####### this function loops over varying sample sizes with n copies per sample size
###### takes argument  1, starting sample size
##### argument 2  , ending sample size
### argument  3, incremental sample size
### argument  4, number of copies per sub-sample
###
##### function returns (A) 4 arrays for h, host-field - indices of significant taxa per subsample per copy, number of significant taxa per subsample per subsample per copy
####                     , h-effect size for detected taxa, L-effect size for detected taxa 
##### function returns (B) 3 arrays for L, abundances - indices of significant taxa per subsample per copy, number of significant taxa per subsample per subsample per copy
####                     , L-effect size for detected taxa 
##### In total, function returns 7 arrays

def analysis_synthetic(n1,n2,inc,ncop):
    ps=0.0
    detections_h = []
    detections_L = []
    Nsig_h=[]
    Nsig_L=[]
    DETS_L=[]
    DETS_H=[]
    nc = 0.01
    effects = []
    effectsH = []
    effectsHfield = []
    samp=[]

    for new_nsamp in range(n1,n2,inc):
        detections_h = []
        detections_L = []

        effs1 = []
        effs1h = []
        effs1hfield = []

        dets_H = []
        dets_L = []
        for copy in range(ncop):
            cd_data = np.load('CD_cop%s_samplesize_%s.npy'% (copy, new_nsamp)) ############# generated CD data
            nsamp_CD = cd_data.shape[0]
            cd_data = ps + cd_data
            cd_data = np.array([(cd_data[i,:])/sum(cd_data[i,:]) for i in range(nsamp_CD)]) ########## normalization - TSS ##########
            cd_data = np.log(cd_data)

            nbact = cd_data.shape[1] ##for gen level
            trunc = nbact
            control_data = np.load('control_cop%s_samplesize_%s.npy'% (copy, new_nsamp)) ############# generated control data
            nsamp_control = control_data.shape[0]
            
            control_data = ps + control_data
            control_data = np.array([(control_data[i,:])/sum(control_data[i,:]) for i in range(nsamp_control)]) ########## normalization - TSS ##########
            control_data =  np.log(control_data)
    
            resamp = np.concatenate((cd_data,control_data), axis = 0)
            cov = np.cov(resamp.T)
            ind1=len(cov)
            U,lambd,Vt = scipy.linalg.svd(cov, full_matrices=True, compute_uv=True, overwrite_a=False, check_finite=True)
            
            for i in range(ind1):
                if lambd[i]<nc:
                    nthresh = i
                    break
      
            lambd_inv = np.zeros(len(cov))
    
            ######### computing inverse singular values #############
            for i in range(nthresh):
                lambd_inv[i] = 1/lambd[i]

            V = Vt.T
            Ut = U.T
            S_inv = scipy.linalg.diagsvd(lambd_inv, ind1, ind1)
            cov_inv = np.dot(V, np.dot(S_inv, Ut))
    
            cov_in = cov_inv
        
            con = control_data
            cd = cd_data
    
            h_con = (np.dot(cov_in,con.T)).T
            h_cd = (np.dot(cov_in,cd.T)).T
    
            samps_l = control_data
            samps_cd_l = cd_data
    
            sampsh = h_con
            samps_cdh = h_cd
            trunc = nbact
    
            p_L = []
            p_h = []

            for i in range(trunc):
            
                tL, pL = ttest_ind(samps_l[:,i], samps_cd_l[:,i], equal_var=True)
                p_L.append(pL)                   

                t, p = ttest_ind(sampsh[:,i], samps_cdh[:,i], equal_var=True)
                p_h.append(p)
        
            con_means = 100*np.exp(np.mean(con, axis = 0))
            cd_means = 100*np.exp(np.mean(cd, axis = 0))
            effs_L = (cd_means/con_means)
            effs_L = [(1.0/x) if x<1.0 else x for x in effs_L]

            hcon_means = (np.mean(h_con, axis = 0))
            hcd_means = (np.mean(h_cd, axis = 0))
            effs_Hfield = (hcd_means-hcon_means)/(np.abs(hcon_means))
            effs_Hfield = np.abs(effs_Hfield)
        
            pick_pvals_L = p_L    
            pick_pvals_h = p_h    
    
            pv_L = np.array(p_L)
            pv_h = np.array(p_h)
    
            pvalsL_allmucosa = pick_pvals_L
            pvalsL_allmucosa.sort()
    
            pvalsh_allmucosa = pick_pvals_h
            pvalsh_allmucosa.sort()

            ranksamp = np.array(pvalsL_allmucosa).shape[0]
            num_total_tests = ranksamp
            prev_bh_valueL = 0
            prev_bh_valueh = 0
            qvalsL_allmucosa= []
            qvalsh_allmucosa= []
                
            for i, (p_value1, p_value2) in enumerate(zip(pvalsL_allmucosa, pvalsh_allmucosa)):
                bh_value = p_value1 * num_total_tests / (i + 1)
                # Sometimes this correction can give values greater than 1,
                # so we set those values at 1
                bh_value = min(bh_value, 1)
                # To preserve monotonicity in the values, we take the
                # maximum of the previous value or this one, so that we
                # don't yield a value less than the previous.
                bh_value = max(bh_value, prev_bh_valueL)
                prev_bh_valueL = bh_value
                qvalsL_allmucosa.append(bh_value)
                
                bh_valueh = p_value2 * num_total_tests / (i + 1)
                bh_valueh = min(bh_valueh, 1)
                bh_valueh = max(bh_valueh, prev_bh_valueh)
                prev_bh_valueh = bh_valueh
                qvalsh_allmucosa.append(bh_valueh)
                
            qvalsL = np.array([qval for qval in qvalsL_allmucosa if qval<0.05])     ############### finding the number of significant taxa with qvalue < 0.05  
            nsigL = qvalsL.shape[0]
            
            qvalsh = np.array([qval for qval in qvalsh_allmucosa if qval<0.05])     ############### finding the number of significant taxa with qvalue < 0.05  
            nsigh = qvalsh.shape[0]
    
            pval_h = pvalsh_allmucosa[0:nsigh]   ############ sorted and significant p values
            pval_L = pvalsL_allmucosa[0:nsigL]   ############ sorted and significant p values

            numm = trunc
            chh_l = []
            for j in range(nsigL):
                dum = pval_L[j]
                for k in range(numm):
                    if pv_L[k] == dum:
                        if k not in chh_l:
                            chh_l.append(k)
       
            chh_h = []
            for j in range(nsigh):
                dum = pval_h[j]
                for k in range(numm):
                    if pv_h[k] == dum:
                        if k not in chh_h:
                            chh_h.append(k)
     
            detections_h.append(nsigh)    
            detections_L.append(nsigL)
            dets_L.append(chh_l)
            dets_H.append(chh_h)
            
            effsL1 = [effs_L[chh_l[k]] for k in range(len(chh_l)) ]
            effs1.append(effsL1) 
            
            effsH1 = [effs_L[chh_h[k]] for k in range(len(chh_h)) ]
            effs1h.append(effsH1) 

            effsH1field = [effs_Hfield[chh_h[k]] for k in range(len(chh_h)) ]
            effs1hfield.append(effsH1field) 
        
        samp.append(new_nsamp)
        effects.append(effs1)
        effectsH.append(effs1h)
        effectsHfield.append(effs1hfield)
        Nsig_L.append(detections_L)
        Nsig_h.append(detections_h)
        DETS_L.append(dets_L)
        DETS_H.append(dets_H)
     
    return np.array(DETS_H), np.array(Nsig_h), np.array(effectsHfield), np.array(effectsH), np.array(DETS_L), np.array(Nsig_L), np.array(effects)
```

# Function 7: computes effect size in new detections as a function of sample size (Figure 2B). Uses output from Function 6

```python
####### this function computes median, mean and minimum effect size for additional taxa detected as a function of sample size 
###### takes arguments from output of function 'synthetic_analysis'
###### takes argument  1, h : host field - indices of significant detections (DETS_H)
###### takes argument  2, h : host field - h-effect size for detected taxa (effectsHfield)
##### takes argument 3 ,  h : host field - L-effect size for detected taxa (effectsH)
##### takes argument 4 ,  L, abundances - indices of significant detections (DETS_L) 
####  takes argument 5,  L, abundances  - L-effect size for detected taxa (effects)
###
##### In total, function returns 3 arrays for Median effect size for new taxa detected as a function of sample size in 
### (A) host-field effects for DAA detected taxa
### (B) L effects for DAA detected taxa
### (C) L effects for MWAS detected taxa 


def effectSizes_samplesize(DETS_H,  effectsHfield, effectsH, DETS_L, effects):
    dets1 = np.array(DETS_L).T
    effs1 = np.array(effects).T
    detsh1 = np.array(DETS_H).T
    effsh1 = np.array(effectsH).T
    effsh1field = np.array(effectsHfield).T

    det = dets1
    eff = effs1
    deth = detsh1
    effh = effsh1
    effhfield = effsh1field

    navg = det.shape[0]
    nparts = det.shape[1]-1
    
    Mean = []
    Median = []
    Min = []

    Meanh = []
    Medianh = []
    Minh = []
    
    Meanhfield = []
    Medianhfield = []
    Minhfield = []
    
    for j in xrange(navg):
        detections_L = det[j]
        effects_L = eff[j]
        detections_H = deth[j]
        effects_H = effh[j]    
        effects_Hfield = effhfield[j]
   
        diffs_mean = []
        diffs_median = []
        diffs_min = []

        diffs_meanh = []
        diffs_medianh = []
        diffs_minh = []

        diffs_meanhfield = []
        diffs_medianhfield = []
        diffs_minhfield = []

        for i in xrange(0,nparts):
            chh1 = detections_L[i]
            chh2 = detections_L[i+1]
            new = [k for k, item in enumerate(chh2) if item not in chh1]
            nsig = len(new)
            etest = effects_L[i+1]
            if not etest: 
                lowest = 0.0
                
            if etest:
                lowest = np.min(etest)
            if nsig ==0:
                effs_new = etest
            if nsig!=0:
                effs_new = ([etest[new[m]] for m in range(nsig)])
            mean = np.mean(effs_new)
            median = np.median(effs_new)
            
            diffs_mean.append(mean)
            diffs_median.append(median)
            diffs_min.append(lowest)
        
            #########################################
        
            chh1 = detections_H[i]
            chh2 = detections_H[i+1]
            new = [k for k, item in enumerate(chh2) if item not in chh1]
            nsig = len(new)
            etest = effects_H[i+1]
            if not etest: 
                lowest = 0.0
            
            if etest:
                lowest = np.min(etest)
            if nsig ==0:
                effs_new = etest
            if nsig!=0:
                effs_new = ([etest[new[m]] for m in range(nsig)])
            mean = np.mean(effs_new)
            median = np.median(effs_new)
            
            diffs_meanh.append(mean)
            diffs_medianh.append(median)
            diffs_minh.append(lowest)

            #########################################
        
            chh1 = detections_H[i]
            chh2 = detections_H[i+1]
            new = [k for k, item in enumerate(chh2) if item not in chh1]
            nsig = len(new)
            etest = effects_Hfield[i+1]
        
            if not etest: 
                lowest = 0.0
            
            if etest:
                lowest = np.min(etest)
            if nsig ==0:
                effs_new = etest
            if nsig!=0:
                effs_new = ([etest[new[m]] for m in range(nsig)])
            mean = np.mean(effs_new)
            median = np.median(effs_new)

            diffs_meanhfield.append(mean)
            diffs_medianhfield.append(median)
            diffs_minhfield.append(lowest)

        Mean.append(diffs_mean)        
        Median.append(diffs_median)        
        Min.append(diffs_min)        

        Meanh.append(diffs_meanh)        
        Medianh.append(diffs_medianh)        
        Minh.append(diffs_minh)        

        Meanhfield.append(diffs_meanhfield)        
        Medianhfield.append(diffs_medianhfield)        
        Minhfield.append(diffs_minhfield)        
  
    Mean1 = np.mean(Mean, axis = 0)
    Median1 = np.mean(Median, axis = 0)
    Min1 = np.mean(Min, axis = 0)

    Mean1h = np.mean(Meanh, axis = 0)
    Median1h = np.mean(Medianh, axis = 0)
    Min1h = np.mean(Minh, axis = 0)

    Mean1hfield = np.mean(Meanhfield, axis = 0)
    Median1hfield = np.mean(Medianhfield, axis = 0)
    Min1hfield = np.mean(Minhfield, axis = 0)

    return Median1hfield,Median1h, Median1
```

# Function for Logistic Regression classifier

```python
#################### LOGISTIC REGRESSION CLASSIFIER ###########
logit = lin.LogisticRegression(penalty='l1', dual=False, tol=0.0001, C=1.0, fit_intercept=True, intercept_scaling=1, class_weight=None, random_state=None, solver='liblinear', max_iter=100, multi_class='ovr', verbose=0, warm_start=False, n_jobs=1)

def logisticclass(data_L , pred_muc, data_H):
    npe=100
    X = data_L
    y = pred_muc
    X_f = data_H
    y_f = y

    nf = 5
    nperm = npe
    perm_auc = np.zeros(nperm)
    perm_auc_f = np.zeros(nperm)

    perm_tpr = 0.0
    perm_tpr_f = 0.0
    perm_fpr = np.linspace(0, 1, 100)
    perm_fpr_f = np.linspace(0, 1, 100)
    means = np.zeros((nperm,100))
    means_f = np.zeros((nperm,100))
    
    ################ repeat k fold cross validation for N permutations ######################
    for j in xrange(nperm):
        kfold = cross_validation.StratifiedKFold(y,shuffle = True, n_folds=nf,random_state=2*j)
        cv = kfold
        classifier = logit
    ############# intialize mean tpr, fpr over each 5-fold cross validation partitioned data-set ############################
    
        mean_tpr = 0.0
        mean_fpr = np.linspace(0, 1, 100)
        mean_tpr_f = 0.0
        mean_fpr_f = np.linspace(0, 1, 100)
    
        for i, (train, test) in enumerate(cv):
    
        # Compute ROC curve and area the curve
            decs_ = classifier.fit(X[train], y[train]).decision_function(X[test])
            fpr, tpr, thresholds = roc_curve(y[test], decs_)

            mean_tpr += interp(mean_fpr, fpr, tpr)
            mean_tpr[0] = 0.0
            roc_auc = auc(fpr, tpr)
    
            # Compute ROC curve and area the curve
            decs_f = classifier.fit(X_f[train], y_f[train]).decision_function(X_f[test])
            fpr_f, tpr_f, thresholds_f = roc_curve(y_f[test], decs_f)
            
            mean_tpr_f += interp(mean_fpr_f, fpr_f, tpr_f)
            mean_tpr_f[0] = 0.0
            roc_auc_f = auc(fpr_f, tpr_f)
        
    ################# calculate mean over 5 folds, ROC curve from mean tpr and fpr ##########################
    
        mean_tpr /= len(cv)
        mean_tpr[-1] = 1.0
        perm_auc[j] = auc(mean_fpr, mean_tpr)
    

        mean_tpr_f /= len(cv)
        mean_tpr_f[-1] = 1.0
        perm_auc_f[j] = auc(mean_fpr_f, mean_tpr_f)
   
     
        means[j] = mean_tpr
        means_f[j] = mean_tpr_f
        
    ####################  record the 5-fold-mean ROC VALUES for the grand curve - SUMMED OVER all permuted partitions   ################### 
        perm_tpr += interp(perm_fpr, mean_fpr, mean_tpr)
        perm_tpr_f += interp(perm_fpr_f, mean_fpr_f, mean_tpr_f)
        
    ############## standard error in mean area under the curve, over all permutations ####################
    ############## standard error in mean area under the curve, over all permutations ####################
    return perm_auc, perm_auc_f, perm_tpr, perm_tpr_f, perm_fpr, perm_fpr_f,  means,  means_f
```
