# Loading the data, filtering out poor statistical data

```python
import numpy as np
import matplotlib.pyplot as plt

############## function to filter out poor statistical data ################
def filter(names,cd_data , control_data, cutoff):
    cd_data = cd_data.T
    control_data = control_data.T
    nbact = cd_data.shape[1]
    nsamp_cd= cd_data.shape[0]
    nsamp_control= control_data.shape[0]
    
    pred =  np.array([[x/x if x > 0 else x for x in cd_data[:,i]] for i in xrange(nbact)])
    pred1 = np.array([np.sum(0.0 + pred[i,:]) for i in xrange(nbact)])/nsamp_CD

    pred =  np.array([[x/x if x > 0 else x for x in control_data[:,i]] for i in xrange(nbact)])
    pred2 = np.array([np.sum(0.0+ pred[i,:]) for i in xrange(nbact)])/nsamp_control

    bad1=[]
    cutoff = 0.38
    for i in xrange(nbact):
        if pred1[i]<cutoff:
            bad1.append(i)
        
    bad2 = []
    for i in xrange(nbact):
        if pred2[i]<cutoff:
            bad2.append(i)

    bad = np.intersect1d(bad1,bad2)
    allbact = np.arange(nbact)

    good = (list(set(allbact) - set(bad)))
    ind1 = len(good)

    names_filt=[]
    for i in xrange(ind1):
        names_filt.append(names[good[i]])

    return good, names_filt
## input to function in the form of data array of shape [ntaxa, nsamples]
def corrected_data(control_data, cd_data, indices):
    nbact = cd_data.shape[0]
    nsamp_cd= cd_data.shape[1]
    nsamp_control= control_data.shape[1]
    CD_data= cd_data.T
    CD_data = 1.0+CD_data ### PSEUDO COUNT #######
    CD_data = np.array([(CD_data[i,:])/sum(CD_data[i,:]) for i in range(nsamp_CD)]) ########## normalization - TSS ##########
    cd_data = np.log(CD_data).T
    
    Control_data = control_data.T
    Control_data = 1.0+Control_data
    Control_data = np.array([(Control_data[i,:])/sum(Control_data[i,:]) for i in range(nsamp_control)]) ########## normalization - TSS ##########
    control_data = np.log(Control_data).T
  
    control= []
    cd = []
    ind1 = np.array(indices).shape[0]
    for i in range(ind1):
        control.append(control_data[indices[i],:])
        cd.append(cd_data[indices[i],:])
 
    return np.array(control), np.array(cd)


```

```python
CD_data = np.load('Raw data files\genus_cd.npy') ############# array of abundances for bacteria in 10 CD stool samples
nbact = len(CD_data)

Control_data = np.load('Raw data files\genus_control.npy') ############# array of abundances for bacteria in 10 CD stool samples

with open ("Raw data files\g__names.txt", "r") as f:
    names= f.read().splitlines()

nsamp_CD = CD_data.shape[1]
nsamp_control = Control_data.shape[1]

good, names_filt_table = filter(names,CD_data, Control_data, 0.38)
control, cd = corrected_data(Control_data, CD_data,good)
control_clus = control.copy()
cd_clus = cd.copy()
good_clus = good
```

# Function for Random Forest Classifier

```python
#################### RANDOM FOREST CLASSIFIER ###########
npe=100
logit = lin.LogisticRegression(penalty='l1', dual=False, tol=0.0001, C=1.0, fit_intercept=True, intercept_scaling=1, class_weight=None, random_state=None, solver='liblinear', max_iter=100, multi_class='ovr', verbose=0, warm_start=False, n_jobs=1)
svmlin = SVC(kernel="linear", C=1.0)
forest = RandomForestClassifier(n_estimators=10, criterion='gini', max_depth=None, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features='auto', max_leaf_nodes=None, bootstrap=True, oob_score=False, n_jobs=1, verbose=0, warm_start=False, class_weight=None)


def forestclass(data_L , pred_muc, data_H):
    npe=100
    X = data_L
    y = pred_muc
    X_f = data_H
    y_f = y


    nf = 5
    nperm = npe
    perm_auc = np.zeros(nperm)
    perm_auc_f = np.zeros(nperm)

    perm_tpr = 0.0
    perm_tpr_f = 0.0
    perm_fpr = np.linspace(0, 1, 100)
    perm_fpr_f = np.linspace(0, 1, 100)
    means = np.zeros((nperm,100))
    means_f = np.zeros((nperm,100))

    ################ repeat k fold cross validation for N permutations ######################
    for j in xrange(nperm):
        kfold = cross_validation.StratifiedKFold(y,shuffle = True, n_folds=nf,random_state=2*j)
        cv = kfold
        classifier = forest
    ############# intialize mean tpr, fpr over each 5-fold cross validation partitioned data-set ############################
        
        mean_tpr = 0.0
        mean_fpr = np.linspace(0, 1, 100)
        mean_tpr_f = 0.0
        mean_fpr_f = np.linspace(0, 1, 100)
    
        for i, (train, test) in enumerate(cv):
    
            # Compute ROC curve and area the curve
            decs_ = classifier.fit(X[train], y[train]).predict_proba(X[test])
            fpr, tpr, thresholds = roc_curve(y[test], decs_[:,1])

            mean_tpr += interp(mean_fpr, fpr, tpr)
            mean_tpr[0] = 0.0
            roc_auc = auc(fpr, tpr)
    
            # Compute ROC curve and area the curve
            decs_f = classifier.fit(X_f[train], y_f[train]).predict_proba(X_f[test])
            fpr_f, tpr_f, thresholds_f = roc_curve(y_f[test], decs_f[:,1])
            
            mean_tpr_f += interp(mean_fpr_f, fpr_f, tpr_f)
            mean_tpr_f[0] = 0.0
            roc_auc_f = auc(fpr_f, tpr_f)
        
    ################# calculate mean over 5 folds, ROC curve from mean tpr and fpr ##########################
    
        mean_tpr /= len(cv)
        mean_tpr[-1] = 1.0
        perm_auc[j] = auc(mean_fpr, mean_tpr)
    
    
        mean_tpr_f /= len(cv)
        mean_tpr_f[-1] = 1.0
        perm_auc_f[j] = auc(mean_fpr_f, mean_tpr_f)
    
        
        means[j] = mean_tpr
        means_f[j] = mean_tpr_f
        
    ####################  record the 5-fold-mean ROC VALUES for the grand curve - SUMMED OVER all permuted partitions   ################### 
        perm_tpr += interp(perm_fpr, mean_fpr, mean_tpr)
        perm_tpr_f += interp(perm_fpr_f, mean_fpr_f, mean_tpr_f)
        


    ############## standard error in mean area under the curve, over all permutations ####################
    return perm_auc, perm_auc_f, perm_tpr, perm_tpr_f, perm_fpr, perm_fpr_f,  means,  means_f


```

```python
def hfield_transform(control , cd, thresh):
    res = np.concatenate((control,cd), axis = 0)
    cov = np.cov(res.T)

    U,lambd,Vt = scipy.linalg.svd(cov, full_matrices=True, compute_uv=True, overwrite_a=False, check_finite=True)
    x = np.arange(len(lambd))
    for i in range(len(cov)):
        cutoff = len(lambd)
        if lambd[i] < thresh:
            cutoff = i
            break

    trunc = len(lambd)
    lambd_inv = np.zeros(trunc)
    ######### computing inverse singular values #############
    for i in range(cutoff):
        lambd_inv[i] = 1.0/lambd[i] 
    V = Vt.T
    Ut = U.T

    S_inv = scipy.linalg.diagsvd(lambd_inv, trunc, trunc)
    ######## inverse covariance matrix #########
    cov_inv = np.dot(V, np.dot(S_inv, Ut))
    h_con1 = np.dot(cov_inv,control.T).T
    h_cd1 = np.dot(cov_inv,cd.T).T
    return lambd, cov_inv, h_con1, h_cd1

#########################################################################################
############################ False discovery rate control ##################

###### takes argument  1, array of h-field p-values, length-(ntaxa), 
##### argument 2 , array of L (abundance) p-values, length-(ntaxa),
### argument  3, level of desired significance alpha, == 0.05 in our analysis
##### function returns (A) for h, host-field - indices of significant taxa, uncorrected pvalues, FDR corrected pvalues
##### function returns (B) for L, abundances - indices of significant taxa, uncorrected pvalues, FDR corrected pvalues
##### In total, function returns 6 arrays


def FDRcontrol(p_h1, p_L1, alpha):
    nmax=1
    nbact = len(p_h1)
    p_h1 = np.array(p_h1).reshape((nmax,nbact))
    p_L1 = np.array(p_L1).reshape((nmax,nbact))
    pick_pvals_L = []
    pick_pvals_h = []

    Nsig_L = []
    Nsig_h = []
    for i in xrange(nmax):
        pick_pvals_h = []
        pick_pvals_L=[]
       ## abundance p values ####
        ph_values = p_h1[i,:].tolist()
        pL_values = p_L1[i,:].tolist()
    
        pick_pvals_h = ph_values
        pick_pvals_L=pL_values
   
   
        ############################################################################################
        pLfilt = np.array(pick_pvals_L).copy()
        phfilt = np.array(pick_pvals_h).copy()
    
        pvalsL_allmucosa = pick_pvals_L
        pvalsh_allmucosa = pick_pvals_h
    
        pvalsL_allmucosa.sort()
        pvalsh_allmucosa.sort()
    
        ranksamp = np.array(pvalsL_allmucosa).shape[0]
        
        num_total_tests = ranksamp
        prev_bh_valueL = 0
        prev_bh_valueh = 0
        qvalsL_allmucosa= []
        qvalsh_allmucosa= []
            
        for i, (p_value1, p_value2) in enumerate(zip(pvalsL_allmucosa, pvalsh_allmucosa)):
            bh_value = p_value1 * num_total_tests / (i + 1)
            # Sometimes this correction can give values greater than 1,
            # so we set those values at 1
            bh_value = min(bh_value, 1)
            # To preserve monotonicity in the values, we take the
            # maximum of the previous value or this one, so that we
            # don't yield a value less than the previous.
            bh_value = max(bh_value, prev_bh_valueL)
            prev_bh_valueL = bh_value
            qvalsL_allmucosa.append(bh_value)
        
            bh_valueh = p_value2 * num_total_tests / (i + 1)
            bh_valueh = min(bh_valueh, 1)
            bh_valueh = max(bh_valueh, prev_bh_valueh)
            prev_bh_valueh = bh_valueh
            qvalsh_allmucosa.append(bh_valueh)
            
        qvalsL = np.array([qval for qval in qvalsL_allmucosa if qval<alpha])     ############### finding the number of significant taxa with qvalue < 0.05  
        nsigL = qvalsL.shape[0]
        qvalsL_all = np.array([qval for qval in qvalsL_allmucosa if qval<1])     ############### finding the number of significant taxa with qvalue < 0.05  
    
    
        qvalsh = np.array([qval for qval in qvalsh_allmucosa if qval<alpha])     ############### finding the number of significant taxa with qvalue < 0.05  
        nsigh = qvalsh.shape[0]
    
        Nsig_h.append(nsigh)
        Nsig_L.append(nsigL)

    print 'number of indirect (for L) significant associations =', np.mean(Nsig_L)
    print 'number of direct (for h) significant associations =', np.mean(Nsig_h)

    pval_h = pvalsh_allmucosa[0:nsigh]   ############ sorted and significant p values
    pval_L = pvalsL_allmucosa[0:nsigL]   ############ sorted and significant p values


    numm = ranksamp
    chh = []
    for i in range(nsigL):
        dum = pval_L[i]
        for j in range(numm):
            if pLfilt[j] == dum:
                if j not in chh:
                    chh.append(j)
    chh_L = np.array(chh)
    
    chh = []
    for i in range(numm):
        dum = pvalsL_allmucosa[i]
        for j in range(numm):
            if pLfilt[j] == dum:
                if j not in chh:
                    chh.append(j)
    chh_all = np.array(chh)
    

    chh=[]
    for i in range(nsigh):
        dum = pval_h[i]
        for j in range(numm):
            if phfilt[j] == dum:
                if j not in chh:
                    chh.append(j)
    chh_h = np.array(chh)
    return chh_h,pval_h, qvalsh, chh_L, pval_L, qvalsL , chh_all,qvalsL_all
```
